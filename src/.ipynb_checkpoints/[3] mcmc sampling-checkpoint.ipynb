# Rejection Sampling
Let's start off by implementing some basic sampling algoritm called `Rejection Sampling`
It uses **Monte Carlo** technicue for sampling it but doenst have **Markov Property**

(The main idea behind it is that it tries to make uniformally map the probability space into a set of desired distributions)

**Here are some wikipedia notes:**

To visualize the motivation behind rejection sampling, imagine graphing the density function of a random variable onto a large rectangular board and throwing darts at it. Assume that the darts are uniformly distributed around the board. Now remove all of the darts that are outside the area under the curve. The remaining darts will be distributed uniformly within the area under the curve, and the x-positions of these darts will be distributed according to the random variable's density. This is because there is the most room for the darts to land where the curve is highest and thus the probability density is greatest.

The visualization as just described is equivalent to a particular form of rejection sampling where the proposal distribution is uniform (hence its graph is a rectangle). The general form of rejection sampling assumes that the board is not necessarily rectangular but is shaped according to some distribution that we know how to sample from (for example, using inversion sampling), and which is at least as high at every point as the distribution we want to sample from, so that the former completely encloses the latter. Otherwise, there will be parts of the curved area we want to sample from that can never be reached. Rejection sampling works as follows:

1. Sample a point on the x-axis from the proposal distribution.
2. Draw a vertical line at this x-position, up to the curve of the proposal distribution.
3. Sample uniformly along this line from 0 to the maximum of the probability density function. If the sampled value is greater than the value of the desired distribution at this vertical line, return to step 1.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats 
from random import choice
%matplotlib inline
```

Let's start off by implementing some sampling algoritms needed to generate our postteriors in much less number of steps

```python
a, b = 3.3, 2.2
x = np.linspace(stats.beta.ppf(0.01, a, b), stats.beta.ppf(0.99, a, b), 100)

fig, ax = plt.subplots(1, 1)
ax.legend('Raw distribution')
ax.plot(x, stats.beta.pdf(x, a, b), 'r-', label='beta pdf')
ax.scatter(x, stats.beta.pdf(x, a, b), c='b', alpha=0.2)
plt.show()
```

```python
def betaMLE(a, b, precision):
     return np.max([stats.beta.pdf(x, a, b) for x in np.linspace(stats.beta.ppf(0.01, a, b), stats.beta.ppf(0.99, a, b), precision)])
    
def betaRejectionSample(a, b, prescision=100, n=10):
    samples = []
    samples_x = []
    x_space = np.linspace(stats.beta.ppf(0.01, a, b), stats.beta.ppf(0.99, a, b), 100)
    mle = betaMLE(a, b, prescision)
    while(len(samples) != n):
        x = choice(x_space)
        y = stats.beta.pdf(x, a, b)
        y_bar = np.random.rand() * mle
        if y > y_bar:
            samples.append(y)
            samples_x.append(x)
        
    samples = [x for _,x in sorted(zip(samples_x, samples))]
    samples_x = list(sorted(samples_x))
    
    return  samples_x, samples
```

```python
samples = betaRejectionSample(a, b, n=20)
```

```python
fig, ax = plt.subplots(1, 1)
ax.legend('Raw distribution')
ax.plot(samples[0], samples[1] , 'r-', label='beta pdf')
ax.scatter(samples[0], samples[1], c='b', alpha=0.2)
plt.show()
```

Now that we know what Rejection Sampling is and how it is working let's discuss why it is silly

- It's costly
- Doesnt work with high dimensional data


But it is cool because it:
- Can directly return independent samples from the distribution
- Free from the problem of autocorrelated samples that is inherent in MCMC methods

The Metropolisâ€“Hastings algorithm works by generating a sequence of sample values in such a way that, as more and more sample values are produced, the distribution of values more closely approximates the desired distribution, P(x). These sample values are produced iteratively, with the distribution of the next sample being dependent only on the current sample value (thus making the sequence of samples into a Markov chain)

```python
# g(p), where p = (x|y), here is symmenrical assuming it is not Hastings
def f_sym(p):
    x, y = p
    return (x-1)**2 + (y-2)**2 - 3**2

def circleMetropolis(a, b, n=100, disc=10):
    samples = []
    samples_x = []
    x_space = np.linspace(stats.beta.ppf(0.01, a, b), stats.beta.ppf(0.99, a, b), 100)
    
    # Here I am using (x, y) representation insted of x[...], y[...]
    r = np.array([0, 0]) # Starting walking from zero
    p = f_sym(r)
    
    i = 0
    while(len(samples) != n):
        r_bar = r + np.random.normal(size=2)
        p_bar = f_sym(r_bar)
        a = np.random.rand() # the acceptance ratio 
        
        if  a < p_bar/p:
            p = p_bar
            r = r_bar
            
            samples.append(r[0])
            samples_x.append(r[1])
                
    samples = [x for _,x in sorted(zip(samples_x, samples))]
    samples_x = list(sorted(samples_x))
        
    return samples_x, samples
```

```python
samples = circleMetropolis(a, b, n=20000)

fig, ax = plt.subplots(1, 1)
ax.legend('Raw distribution')
ax.plot(samples[0], samples[1] , 'r-', label='beta pdf')
ax.scatter(samples[0], samples[1], c='b', alpha=0.2)
plt.show()
```
